```python
import pandas as pd
import os
import glob

free_trade_path = os.path.join(base_path, "data/raw/free_trade_export_import")

files = glob.glob(os.path.join(free_trade_path, "*20231231*.csv"))

print("선택된 파일 수:", len(files))
for f in files:
    print(os.path.basename(f))
```

```python
# 자유무역지역 파일 불러오기

free_trade_file = files[0]

free_trade = pd.read_csv(free_trade_file, encoding="cp949")

print("데이터 크기:", free_trade.shape)
display(free_trade.head())
print(free_trade.columns)
```

```python
# 마산 자유무역지역만 추출

masan_free_trade = free_trade[
    [
        "연도",
        "마산_수출실적(천달러)",
        "마산_수입실적(천달러)",
        "마산_무역수지(천달러)",
        "마산_고용인원",
        "마산_업체수"
    ]
].copy()

display(masan_free_trade.tail())
```

원래는 22-24년도 데이터 보고 싶은데 현재 파일은 23년까지 있기 때문에 20~23 추세 + 22~23 최근 변화 사용

```python
masan_recent = masan_free_trade[
    masan_free_trade["연도"].between(2020, 2023)
].copy()

display(masan_recent)
```

```python
# 총교역액 만들기

masan_recent["마산_총교역액(천달러)"] = (
    masan_recent["마산_수출실적(천달러)"] +
    masan_recent["마산_수입실적(천달러)"]
)

display(masan_recent)


#################################################

# 전년 대비 증가율 계산

masan_recent = masan_recent.sort_values("연도").reset_index(drop=True)

growth_cols = [
    "마산_수출실적(천달러)",
    "마산_수입실적(천달러)",
    "마산_총교역액(천달러)",
    "마산_고용인원",
    "마산_업체수"
]

for col in growth_cols:
    masan_recent[col + "_전년대비증가율"] = masan_recent[col].pct_change() * 100

display(masan_recent)
```

2020년 기준 지수 만들기 - 20년 대비 얼마나 커졋냐 보는 거

```python
base_year = 2020

for col in growth_cols:
    base_value = masan_recent.loc[
        masan_recent["연도"] == base_year, col
    ].values[0]

    masan_recent[col + "_2020기준지수"] = (
        masan_recent[col] / base_value * 100
    )

display(masan_recent)
```

핵심결과해석
1) 총 교역액 증가 (금액기준임, 양이 증가한 것은 아님), 23년에는 마산 자유무욕지역 총교역액은 20년 대비 약 34.0% 증가
2) 수출실적도 꾸준하게 증가, 23년 수출실적은 20년 대비 약 37.4% 증가 -> 마산 지유무역 지역에서 수출 기반 산업 활동이 확대되고 있단 근거로 사용 가능
3) 수입실적은 2022년에 잠깐 줄었지만 2023년에 다시 증가, 2022년에 전년 대비 감소하였지만 2023년에는 다시 8.15% 증가하여 회복 흐름보임
4) 업체 수 증가, 23년 업체수는 2020년 대비 약 9.2% 증가 -> 물류를 발생시키는 기업 수 늘은 의미로 볼 수 있음

=> 마산 자유무역지역은 20~23년 동안 수출실적, 총교역액, 업체 수 모두 증가, 이는 창원항만 배후 산업권역에서 수출입 기반 물류 수요가 지속적으로 발생하고 확대되고 있음을 보여주는 간접 지표로 해석 가능
