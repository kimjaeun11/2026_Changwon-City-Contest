```python
import glob
import os

port_facility_path = os.path.join(base_path, "data/raw/port_facility")

files = glob.glob(os.path.join(port_facility_path, "*.csv"))

print("전체 파일 수:", len(files))

for f in files:
    print(os.path.basename(f))
```

21, 22, 24 파일만 선택

```python
target_years = ["2021", "2022", "2024"]

selected_files = [
    f for f in files
    if any(year in os.path.basename(f) for year in target_years)
]

print("선택된 파일 수:", len(selected_files))

for f in selected_files:
    print(os.path.basename(f))
```

파일 불러오고 통합

```python
import pandas as pd
import re

df_list = []

for file in selected_files:
    df = pd.read_csv(file, encoding="utf-8-sig")

    year_match = re.search(r"20\d{2}", os.path.basename(file))
    if year_match:
        df["year"] = int(year_match.group())

    df["source_file"] = os.path.basename(file)
    df_list.append(df)

port_facility = pd.concat(df_list, ignore_index=True)

print("통합 데이터 크기:", port_facility.shape)
display(port_facility.head())
print(port_facility.columns)
```

```python
print(port_facility["year"].value_counts().sort_index())  # 연도 확
```

```python
print(port_facility["prtAgNm"].unique()) # 항만명 확인
```

마산항, 진해항만 필터링

```python
target_ports = ["마산", "진해"]

port_facility_changwon = port_facility[
    port_facility["prtAgNm"].astype(str).str.contains("|".join(target_ports), na=False)
].copy()

print("마산·진해 항만시설 데이터 크기:", port_facility_changwon.shape)

print("항만명:")
print(port_facility_changwon["prtAgNm"].unique())

print("연도별 데이터 수:")
print(port_facility_changwon["year"].value_counts().sort_index())

display(port_facility_changwon.head())
```

하역능력 컬럼 숫자형 정리

```python
port_facility_changwon["lnlAblty"] = (
    port_facility_changwon["lnlAblty"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace("-", "0", regex=False)
    .str.strip()
)

port_facility_changwon["lnlAblty"] = pd.to_numeric(
    port_facility_changwon["lnlAblty"],
    errors="coerce"
).fillna(0)
```

```python
lnl_check = (
    port_facility_changwon
    .groupby(["year", "prtAgNm"], as_index=False)
    .agg(
        facility_count=("fcltyKorNm", "count"),
        lnlAblty_sum=("lnlAblty", "sum"),
        lnlAblty_nonzero=("lnlAblty", lambda x: (x > 0).sum())
    )
)

display(lnl_check)
```

lnlAblty_nonzero가 거의 0이면, 이 데이터에서는 하역능력 수치 분석보다 시설 수·시설명 중심 분석으로 가야

```python
# 연도별 항만별 시설 수 집계

facility_yearly = (
    port_facility_changwon
    .groupby(["year", "prtAgNm"], as_index=False)
    .agg(
        facility_count=("fcltyKorNm", "count"),
        unique_facility_code=("fcltyCd", "nunique"),
        unique_facility_name=("fcltyKorNm", "nunique"),
        total_lnlAblty=("lnlAblty", "sum")
    )
)

display(facility_yearly)
```

2021, 2022, 2024년에 마산항과 진해항의 항만시설 항목 수가 얼마나 되는지 확인할 수 있음

```python
# 시설 구분별 집계

facility_type_summary = (
    port_facility_changwon
    .groupby(["year", "prtAgNm", "useFeeKorNm"], as_index=False)
    .size()
    .rename(columns={"size": "count"})
)

display(facility_type_summary)
```

useFeeKorNm은 접안료, 정박료 같은 시설/사용료 구분 - 마산항·진해항 시설이 접안, 정박, 하역 등 어떤 기능과 관련되어 있는지 확인할 수 있음

```python
# 마산항 시설명

masan_facility_names = (
    port_facility_changwon[port_facility_changwon["prtAgNm"] == "마산"]
    [["year", "fcltyKorNm", "useFeeKorNm", "lnlDvicNm", "lnlAblty", "trtmntFrghtNm", "berthNm"]]
    .drop_duplicates()
)

display(masan_facility_names.head(50))
```

```python
# 진해항 시설명


jinhae_facility_names = (
    port_facility_changwon[port_facility_changwon["prtAgNm"] == "진해"]
    [["year", "fcltyKorNm", "useFeeKorNm", "lnlDvicNm", "lnlAblty", "trtmntFrghtNm", "berthNm"]]
    .drop_duplicates()
)

display(jinhae_facility_names.head(50))
```

이 표는 보고서에서 *대표 시설 예시*로 쓰면 좋을듯

```python
# 취급화물명 확인

cargo_summary = (
    port_facility_changwon
    .dropna(subset=["trtmntFrghtNm"])
    .groupby(["year", "prtAgNm", "trtmntFrghtNm"], as_index=False)
    .size()
    .rename(columns={"size": "count"})
)

display(cargo_summary.head(50))
```

취급화물명 확인 결과는 No entries - port_facility_changwon 안에서 trtmntFrghtNm, 즉 취급화물명 컬럼이 마산·진해 데이터에서는 비어 있음
-> 마산항은 컨테이너 화물을 주로 취급 / 진해항은 벌크화물을 주로 취급

```python
# 선석명 확인

berth_summary = (
    port_facility_changwon
    .dropna(subset=["berthNm"])
    .groupby(["year", "prtAgNm", "berthNm"], as_index=False)
    .size()
    .rename(columns={"size": "count"})
)

display(berth_summary.head(50))
```

count가 실제 선석 개수를 의미하는 건 아님
같은 선석이 여러 시설 항목에 반복해서 나올 수 있어서 count는 등록 횟수에 가까움
마산항과 진해항에 어떤 선석 또는 접안시설 명칭이 등록되어 있는지 확인하는 보조자료로 활용

취급화물명: 값이 없어 분석 제외
선석명: 항만시설 위치·접안시설 명칭 확인용으로 활용
하역능력: 0 또는 결측이 많으면 총량 비교는 제한
핵심 분석: 시설 수, 시설명, 사용료 구분, 선석명 중심

이 데이터의 최종 역할은: 마산항과 진해항의 항만시설 기반이 존재한다는 것을 확인하는 자료

```python
processed_path = os.path.join(base_path, "data/processed")
os.makedirs(processed_path, exist_ok=True)

facility_yearly.to_csv(
    os.path.join(processed_path, "changwon_port_facility_yearly_2021_2022_2024.csv"),
    index=False,
    encoding="utf-8-sig"
)

facility_type_summary.to_csv(
    os.path.join(processed_path, "changwon_port_facility_type_summary_2021_2022_2024.csv"),
    index=False,
    encoding="utf-8-sig"
)

cargo_summary.to_csv(
    os.path.join(processed_path, "changwon_port_facility_cargo_summary_2021_2022_2024.csv"),
    index=False,
    encoding="utf-8-sig"
)

berth_summary.to_csv(
    os.path.join(processed_path, "changwon_port_facility_berth_summary_2021_2022_2024.csv"),
    index=False,
    encoding="utf-8-sig"
)

port_facility_changwon.to_csv(
    os.path.join(processed_path, "changwon_port_facility_filtered_2021_2022_2024.csv"),
    index=False,
    encoding="utf-8-sig"
)

print("항만시설 및 하역능력정보 2021, 2022, 2024 분석 결과 저장 완료")
```

이 분석은 마산항·진해항의 항만시설 기반이 실제로 존재하는지 확인하기 위한 단계이고, 결론은 “창원 항만권역에는 항만시설 기반이 확인되지만, 하역능력·취급화물 정보가 부족해 시설 현황 중심으로만 활용해야 한다

```python
# 부산 시설 구분 확

busan_facility_type_summary = (
    port_facility_busan
    .groupby(["year", "prtAgNm", "useFeeKorNm"], as_index=False)
    .size()
    .rename(columns={"size": "count"})
)

display(busan_facility_type_summary)
```

```python
# 부산 선석명 확인

busan_berth_summary = (
    port_facility_busan
    .dropna(subset=["berthNm"])
    .groupby(["year", "prtAgNm", "berthNm"], as_index=False)
    .size()
    .rename(columns={"size": "count"})
)

display(busan_berth_summary.head(50))
```

```python
# 선석 종류 개수만 보고 싶다면

busan_berth_count_summary = (
    port_facility_busan
    .dropna(subset=["berthNm"])
    .groupby(["year", "prtAgNm"], as_index=False)
    .agg(
        unique_berth_count=("berthNm", "nunique")
    )
)

display(busan_berth_count_summary)
```

```python
# 부산 취급화물명 확인

busan_cargo_summary = (
    port_facility_busan
    .dropna(subset=["trtmntFrghtNm"])
    .groupby(["year", "prtAgNm", "trtmntFrghtNm"], as_index=False)
    .size()
    .rename(columns={"size": "count"})
)

display(busan_cargo_summary.head(50))
```

마산 진해 vs 부산 비교표 만들기 - 부산을 비교 기준으로 잡으면, 창원 항만권역(마산 진해) 이 상대적으로 어느 정도 규모인지 설명하기 좋음

```python
compare_ports = ["마산", "진해", "부산"]

port_facility_compare = port_facility[
    port_facility["prtAgNm"].astype(str).str.contains("|".join(compare_ports), na=False)
].copy()

port_facility_compare["lnlAblty"] = (
    port_facility_compare["lnlAblty"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace("-", "0", regex=False)
    .str.strip()
)

port_facility_compare["lnlAblty"] = pd.to_numeric(
    port_facility_compare["lnlAblty"],
    errors="coerce"
).fillna(0)

facility_compare_summary = (
    port_facility_compare
    .groupby(["year", "prtAgNm"], as_index=False)
    .agg(
        facility_count=("fcltyKorNm", "count"),
        unique_facility_name=("fcltyKorNm", "nunique"),
        unique_berth_count=("berthNm", "nunique"),
        lnlAblty_sum=("lnlAblty", "sum"),
        lnlAblty_nonzero=("lnlAblty", lambda x: (x > 0).sum())
    )
)

display(facility_compare_summary)
```

부산항은 마산항·진해항보다 항만시설 항목 수와 선석 정보가 훨씬 많이 확인된다. 반면 마산항과 진해항도 시설 기반은 존재하지만, 부산항에 비해 상대적으로 제한적인 규모로 나타남

- 부산항 시설 규모가 압도적으로 크다
- 마산항은 21, 22, 24년 시설 항목 수 일정하게 유지되어서 안정적인 항만시설 기반을 유지함
- 진해항은 21년도보다 22,24년도에 많아짐 (실제 시설 확충인지, 데이터 등록 방식 변화인지는 확인이 필요하니까 '증가한 것으로 확인된다' 정도로만 쓰기)
- unique_berth_count (선석 정보)는 부산만 확인됨 -> 마산 진해항에는 선석이 없다는 것이 아니라 마산항과 진해항 자료에서는 선석명 정보가 제공되지 않았고, 부산항 자료에서는 48~49개의 고유 선석명이 확인됨, 부산항은 선석명 정보도 구체적으로 제공되지만, 마산 진해는 선석명 데이터가 비어있는 것
- nlAblty_sum이 전부 0이고, lnlAblty_nonzero도 전부 0 -> 21, 22, 24년 마산 진해 부산 항만시설 자료에서 하역능력 수치가 유효하게 제공되지 않았음

=> 보고서 작성시 하역능력 수치가 0으로 제공되어 항만별 하역능력 총량 비교에는 한계가 있다. 따라서 본 분석에서는 시설 항목 수와 시설명, 선석명 제공 여부를 중심으로 항만시설 기반을 비교하였다.  이런식으로 작성


### 부산항은 마산항·진해항에 비해 시설 항목 수와 선석 정보가 훨씬 많이 확인되어 동남권 대표 항만으로서 큰 항만시설 기반을 가짐, 반면 마산항과 진해항도 항만시설 기반은 확인되지만, 부산항과 비교하면 상대적으로 제한적인 규모로 나타남. 따라서 창원 항만권역은 독립적인 대형 항만 물류 중심지라기보다는 부산항과 연계되는 배후 물류권역으로 해석 가능
