```python
import pandas as pd
broker=pd.read_csv('경상남도 창원시_화물운송 주선업체 현황_20230516.csv', encoding='CP949')
broker.head()
```

```python
print(f"데이터 크기: {broker.shape}")
print(f"비고 NaN 개수: {broker['비고'].isna().sum()}")
print(f"비고 NaN 비율: {broker['비고'].isna().sum() / len(broker) * 100:.1f}%")
broker = broker.drop(columns=['비고'])
broker.head()
```

```python
# 구별 업체 수
print(broker['해당구'].value_counts())
```

```python
# 구별 집계
gu_broker = broker.groupby('해당구').size().reset_index()
gu_broker.columns = ['구', '주선업체수']
print(gu_broker.sort_values('주선업체수', ascending=False))
```

```python
# 진해구 vs 나머지
broker['진해여부'] = broker['해당구'].apply(lambda x: '진해구' if x == '진해구' else '기타')
print(broker.groupby('진해여부').size())
```

```python
#운송업체 수 대비 주선업체 비율이 진해구가 제일 낮아요. 이게 공모전 핵심 논거로 쓸 수 있어요.
#"진해구는 운송업체 43개 대비 주선업체 9개로 비율 0.21 — 창원시 최하위. 화물-차량 연결 인프라가 가장 취약한 지역임에도 신항이 위치."
```
