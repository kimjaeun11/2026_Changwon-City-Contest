```python
# 구별 운송사업자 및 차량대수

import pandas as pd
import os

base_path = "/content/drive/MyDrive/contest_changwon_port"

trucking_path = os.path.join(
    base_path,
    "data/raw/trucking_company",
    "경상남도 창원시_일반화물자동차운송사업자_20230620.csv"
)

trucking = pd.read_csv(trucking_path, encoding="cp949")

print("데이터 크기:", trucking.shape)
display(trucking.head())
print(trucking.columns)
```

```python
# 차량대수 숫자 처리

vehicle_cols = [
    "차량대수(직영)",
    "차량대수(위수탁)",
    "차량대수(확인불가)"
]

for col in vehicle_cols:
    trucking[col] = pd.to_numeric(trucking[col], errors="coerce").fillna(0)

trucking["총차량대수"] = (
    trucking["차량대수(직영)"] +
    trucking["차량대수(위수탁)"] +
    trucking["차량대수(확인불가)"]
)

display(trucking.head())
```

```python
trucking["구"] = trucking["주소"].astype(str).str.extract(r"창원시\s*([^\s]+구)")[0]

print(trucking["구"].unique())
display(trucking[["업체명", "주소", "구"]].head())
```

```python
# 구별 업체 수, 차량대수 집계

trucking_gu_summary = (
    trucking
    .groupby("구", as_index=False)
    .agg(
        업체수=("업체명", "count"),
        직영차량대수=("차량대수(직영)", "sum"),
        위수탁차량대수=("차량대수(위수탁)", "sum"),
        확인불가차량대수=("차량대수(확인불가)", "sum"),
        총차량대수=("총차량대수", "sum")
    )
)

trucking_gu_summary["업체비중"] = (
    trucking_gu_summary["업체수"] / trucking_gu_summary["업체수"].sum() * 100
)

trucking_gu_summary["차량비중"] = (
    trucking_gu_summary["총차량대수"] / trucking_gu_summary["총차량대수"].sum() * 100
)

trucking_gu_summary["업체당_평균차량대수"] = (
    trucking_gu_summary["총차량대수"] / trucking_gu_summary["업체수"]
)

trucking_gu_summary = trucking_gu_summary.sort_values(
    "총차량대수",
    ascending=False
)

display(trucking_gu_summary)
```

```python
# 항만 인접지역 / 내륙 산단 배후 지역 비교

def classify_area(gu):
    if gu in ["마산합포구", "진해구"]:
        return "항만 인접지역"
    else:
        return "내륙·산단 배후지역"

trucking["권역구분"] = trucking["구"].apply(classify_area)

trucking_area_summary = (
    trucking
    .groupby("권역구분", as_index=False)
    .agg(
        업체수=("업체명", "count"),
        총차량대수=("총차량대수", "sum")
    )
)

trucking_area_summary["업체비중"] = (
    trucking_area_summary["업체수"] / trucking_area_summary["업체수"].sum() * 100
)

trucking_area_summary["차량비중"] = (
    trucking_area_summary["총차량대수"] / trucking_area_summary["총차량대수"].sum() * 100
)

trucking_area_summary["업체당_평균차량대수"] = (
    trucking_area_summary["총차량대수"] / trucking_area_summary["업체수"]
)

display(trucking_area_summary)
```

```python
license_summary = (
    trucking
    .groupby(["구", "면허종류"], as_index=False)
    .agg(
        업체수=("업체명", "count"),
        총차량대수=("총차량대수", "sum")
    )
    .sort_values(["구", "총차량대수"], ascending=[True, False])
)

display(license_summary)
```

```python
processed_path = os.path.join(base_path, "data/processed")
os.makedirs(processed_path, exist_ok=True)

trucking_gu_summary.to_csv(
    os.path.join(processed_path, "trucking_gu_summary_2023.csv"),
    index=False,
    encoding="utf-8-sig"
)

trucking_area_summary.to_csv(
    os.path.join(processed_path, "trucking_area_summary_2023.csv"),
    index=False,
    encoding="utf-8-sig"
)

license_summary.to_csv(
    os.path.join(processed_path, "trucking_license_summary_2023.csv"),
    index=False,
    encoding="utf-8-sig"
)

trucking.to_csv(
    os.path.join(processed_path, "trucking_processed_2023.csv"),
    index=False,
    encoding="utf-8-sig"
)

print("일반화물자동차운송사업자 분석 결과 저장 완료")
```

일반화물자동차운송사업자 현황 분석 결과, 창원시에는 총 566개의 일반화물자동차 운송사업자가 확인, 구별로는 진해구 135개, 의창구 134개, 마산회원구 121개, 성산구 103개, 마산합포구 73개 순으로 나타남, 그러나 차량대수 기준으로는 마산회원구가 1,068대로 가장 많았고, 의창구 887대, 진해구 580대, 성산구 506대, 마산합포구 453대 순으로 확인
이는 업체 수와 실제 운송 공급력이 반드시 일치하지 않음을 보여줌


항만 인접지역인 마산합포구와 진해구에는 총 208개 업체와 1,033대의 차량이 분포한 반면, 의창구·성산구·마산회원구 등 내륙·산단 배후지역에는 358개 업체와 2,461대의 차량이 분포, 따라서 창원시의 화물차 운송 공급력은 항만 인접지역보다 내륙 및 산업 배후지역에 더 집중되어 있는 것으로 나타남, 이는 향후 항만과 산업단지, 운송사업자 간 공차 이동을 줄이기 위한 귀로 화물 매칭 모델 구축의 필요성을 뒷받침함
